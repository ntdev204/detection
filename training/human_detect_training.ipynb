```
from ultralytics import YOLO
import os
import glob
import shutil

dataset_path = "/content/gdrive/My Drive/Colab Notebooks/Deep-Detection/Human"
data_yaml = f"{dataset_path}/data.yaml"

print(f"Dataset location: {dataset_path}")
print(f"Data YAML: {data_yaml}")
print(f"Data YAML exists: {os.path.exists(data_yaml)}")
```

```
model = YOLO("yolov8n.pt")
```

```
results = model.train(
    data=data_yaml,
    epochs=100,
    imgsz=640,
    batch=16,
    name="human-detection",
    patience=20,
    save=True,
    device=0,
    project="runs/detect",
    exist_ok=True,
    pretrained=True
)
```

```
best_paths = glob.glob("runs/detect/**/best.pt", recursive=True)
if best_paths:
    latest_best = max(best_paths, key=os.path.getmtime)
    print(f"Found best.pt at: {latest_best}")
    shutil.copy(latest_best, "/content/best.pt")
    print("Copied to /content/best.pt")
else:
    print("best.pt not found. Check if training completed successfully.")
    !ls -R runs/detect/
```
